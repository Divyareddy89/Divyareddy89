# merge and joins

```python
# create sample data
data1 = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df1 = pd.DataFrame(data1, index=['a', 'b', 'c'])

data2 = {'C': [10, 20, 30], 'D': [40, 50, 60]}
df2 = pd.DataFrame(data2, index=['a', 'c', 'd'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
data1 = {'A': [1, 2, 3], 'B': [4, 5, 6], 'key': ['a', 'b', 'c']}
df1 = pd.DataFrame(data1)

data2 = {'C': [10, 20, 30], 'D': [40, 50, 60], 'key': ['a', 'c', 'd']}
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,left_index=False, right_index=False, sort=True)
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on="id")
```

```python
pd.merge(left,right,on="subject_id")
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=["id","subject_id"])
```

```python
pd.merge(left, right, on='subject_id', how='left')
```

```python
pd.merge(left, right, on='subject_id', how='right')
```

```python
pd.merge(left, right, on='subject_id', how='outer')
```
